### 8. Pабота с пропущенными значениями

Из статистики становится видно, что пропущенные значения имеются в столбцах "Year" (Год) и "Publisher" (Издатель).<br>
Данные пропуски заполним следующим образом:
* значения в столбце "Year" (Год) заполним средним значением в столбце: 2006 год.
* значения в столбце "Publisher" (Издатель) заполним значеним "Wargaming"

### 9. Cоздание нового поля вычисленного на основе значений других полей.

Перед выполнением данного задания, произведем небольшую корректировку DataFrame: вещественные показатели продаж приведем к единичным экземплярам, поскольку на данный момент они исчисляются в миллионах экземпляров.

```python
data.loc[data["NA_Sales"] > 0, "NA_Sales"]=data["NA_Sales"]*1000000;
data.loc[data["EU_Sales"] > 0, "EU_Sales"]=data["EU_Sales"]*1000000;
data.loc[data["JP_Sales"] > 0, "JP_Sales"]=data["JP_Sales"]*1000000;
data.loc[data["Other_Sales"] > 0, "Other_Sales"]=data["Other_Sales"]*1000000;
data.loc[data["Global_Sales"] > 0, "Global_Sales"]=data["Global_Sales"]*1000000;
```

#### 9.1 Через выражение на базе имеющихся колонок.

Создадим несколько новых колонок, на без имеющихся колонок.<br>
На основе столбцов отображающих, количество копий проданных в различных регионах, в том числе и по всему миру,<br>cоздадим новые которые бы, отражали возможную прибыль при продаже такого количества копий игр.

```python
data["NA_Sales_Dollars"] = data["NA_Sales"]*np.random.randint(1, 100);
data["EU_Sales_Dollars"] = data["EU_Sales"]*np.random.randint(1, 100);
data["JP_Sales_Dollars"] = data["JP_Sales"]*np.random.randint(1, 100);
data["Other_Sales_Dollars"] = data["Other_Sales"]*np.random.randint(1, 100);
data["Global_Sales_Dollars"] = data["Global_Sales"]*np.random.randint(1, 100);
```

Просмотрим результат

```python
data
```

#### 9.2 Через DataFrame.apply

Создадим новую колонку с данными используя DataFrame.apply, которая будет говорить нам о прибыльности той или иной игры:<br>
Игра является прибыльной в случае, если ее мировые сборы превысят 50 млн долларов США.

```python
data["Money_Status"] = data.apply(lambda x: "Profitable" if x["Global_Sales_Dollars"] > 50000000 else "Not profitable", axis=1)
```

Проверим результат.

```python
data.sample(10)
```

#### 9.3 через Series.apply

Создадим новую колонку с данными используя Series.apply, которая будет говорить нам о принадлежности той или иной игры к поколению игр:<br>
Игра является игрой нового поколения в случае, если она выпущена была после 2007 года.

```python
data["Status_Gen"] = data["Year"].apply(lambda x: "New Gen" if x > 2007 else "Old Gen")
```

## Этап 4. Решейпинг данных 1Dto2D с помощью .pivot.

Решейпинг данных 1Dto2D с помощью .pivot (можно подать на вход результаты агрегатов, полученных ранее через .groupby (сгруппировать по двум полям), либо прекрасно заходит сюда данные из SQL сгруппированные предварительно по 2-м полям).
